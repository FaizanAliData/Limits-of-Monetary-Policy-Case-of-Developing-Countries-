 Milton Friedman in his presidential address said that '**Inflation is always and everywhere a monetary phenomenon**'. This study focuses on the empirical evaluation of Friedman's money supply growth rule equation and validation of inflation as a purely monetary phenomenon in developing economies. As a case study, the research employs data from Pakistan, India, Sri Lanka, and Bangladesh to test this theory.

The study is based on the following research objectives:


1.   Finding the Co-integration between P (CPI inflation) and P*(Inflation due to excess money supply/Actual inflation).

2.   Finding the Co-integration between the interest rate (policy rate) and Inflation (P) and analysing the granger causality between the two.

3.   If (1) does not yield unity, analysing the impact of other factors (lag of CPI, Money Supply growth rate, GDP growth rate, Exhange rate, Food price, interest rate) on inflation (P).   





The study employs 40 year datasets from the periods 1980-2020 from the national sources of each country such as the statistical handbooks and the central banks. (All the values are in percentage changes)

1. Finding the Co-integration between P and P*.

First **calculate the growth rates from money supply and real GDP of each country**. The difference between the Money Supply Growth rate and the real GDP growth rate will return the **value of P_star** (excess of money supply).

Finally, **co-integrate P and P_star** to find the relationship between the two.

*Note : A one to one relationship shows that inflation is a purely monetary formula while any deviation from that shows that inflation is also impacted by non-monetary factors*

```python
## Calculating the Money supply growth rates
M2_gr_pak=df_pak['M2_pak'].pct_change()*100
M2_gr_ind=df_ind['M2_ind'].pct_change()*100
M2_gr_sri=df_sri['M2_sri'].pct_change()*100
M2_gr_bang=df_bang['M2_bang'].pct_change()*100

#print(M2_gr_pak)
#print(M2_gr_ind)
#print(M2_gr_sri)
#print(M2_gr_ban)
```

```python
## Calculating the real GDP growth rates
RGDP_gr_pak=df_pak['RGDP_pak'].pct_change()*100
RGDP_gr_ind=df_ind['RGDP_ind'].pct_change()*100
RGDP_gr_sri=df_sri['RGDP_sri'].pct_change()*100
RGDP_gr_bang=df_bang['RGDP_bang'].pct_change()*100

#print(RGDP_gr_pak)
#print(RGDP_gr_ind)
#print(RGDP_gr_sri)
#print(RGDP_gr_ban)
```

```python
#Adding the calculated values to the dataset

##M2 growth rates to the dataset
merged_df['M2_gr_pak']=M2_gr_pak
merged_df['M2_gr_ind']=M2_gr_ind
merged_df['M2_gr_sri']=M2_gr_sri
merged_df['M2_gr_bang']=M2_gr_bang



##RGDP growth rates to the dataset
merged_df['RGDP_gr_pak']=RGDP_gr_pak
merged_df['RGDP_gr_ind']=RGDP_gr_ind
merged_df['RGDP_gr_sri']=RGDP_gr_sri
merged_df['RGDP_gr_bang']=RGDP_gr_bang

merged_df.columns

```

```python
# Calculating P* and adding it to the dataset
P_star_pak =merged_df['M2_gr_pak']-merged_df['RGDP_gr_pak']
P_star_ind =merged_df['M2_gr_ind']-merged_df['RGDP_gr_ind']
P_star_sri =merged_df['M2_gr_sri']-merged_df['RGDP_gr_sri']
P_star_bang =merged_df['M2_gr_bang']-merged_df['RGDP_gr_bang']

## Adding P_star values to the dataset
merged_df['P_star_pak']=P_star_pak
merged_df['P_star_ind']=P_star_ind
merged_df['P_star_sri']=P_star_sri
merged_df['P_star_bang']=P_star_bang

#merged_df.head()
#merged_df.tail()
```

The regression results indicate that P and P_star do not have a perfect unit r/s b/w them. However, the results are isgnificant at 5% significance level. To overcome this issue we will **regress P with the lag of P_star** to check if it imporves the significance of the result.

2. Finding the Co-integration between the interest rate (policy rate) and Inflation (P) and analysing the granger causality between the two.

First we find the relationship between interest rate and inflation by running a simple OLS regression.

```python
# Finding the co-integration between the interest rate and inflation.

## For Pakistan
P_pak = merged_df['CPI_pak']
r_pak = merged_df['IR_pak']



#adding constant to predictor variable
r_pak = sm.add_constant(r_pak)

#fitting linear regression model
model = sm.OLS(P_pak, r_pak, missing='drop').fit()

print(model.summary())
```

```python
## For India
P_ind = merged_df['CPI_ind']
r_ind = merged_df['IR_ind']



#adding constant to predictor variable
r_ind = sm.add_constant(r_ind)

#fitting linear regression model
model = sm.OLS(P_ind, r_ind, missing='drop').fit()

print(model.summary())
```

```python
#For Bangladesh
P_bang = merged_df['CPI_bang']
r_bang = merged_df['IR_bang']



#adding constant to predictor variable
r_bang = sm.add_constant(r_bang)

#fitting linear regression model
model = sm.OLS(P_bang, r_bang, missing='drop').fit()

print(model.summary())
```

```python
#For Srilanka
P_sri = merged_df['CPI_sri']
r_sri = merged_df['IR_sri']



#adding constant to predictor variable
r_sri = sm.add_constant(r_sri)

#fitting linear regression model
model = sm.OLS(P_sri, r_sri, missing='drop').fit()

print(model.summary())
```

Thus, since there is a **positive r/s between inflation and interest rate**, the text book approach of increasing interest rate to curb inflation is actually increasing the inflation for the case of these four developing countries at least.

*Note: All the results are significant at 5% except Pakistan which is significant at 10%*

Next we run Granger Casuality test to see whether inflation causes interest rate to rise or whether an increase in interest rate increases inflation.

```python
#performing Granger-Causality test
from statsmodels.tsa.stattools import grangercausalitytests

#For Pakistan
grangercausalitytests(merged_df[['CPI_pak', 'IR_pak']].fillna(0), maxlag=[3])
```

Thus, we see that interest rate causes inflation to rise (p-value is less than 0.05). Now, let's see if the opposite is true.

```python
grangercausalitytests(merged_df[['IR_pak', 'CPI_pak']].fillna(0), maxlag=[3])
```

This shows that inflation does not cause interest rate ti rise for the case of Pakistan (i.e p-value is greater than 0.05). Similarly we will do this for all the countries

```python
#For India
grangercausalitytests(merged_df[['CPI_ind', 'IR_ind']].fillna(0), maxlag=[3])
```

```python
grangercausalitytests(merged_df[['IR_ind', 'CPI_ind']].fillna(0), maxlag=[3])
```

For the case of India, the causality is also from the side of the interest rate

3. Analysing the impact of other factors (lag of CPI, Money Supply growth rate, GDP growth rate, Exhange rate, Food price, interest rate) on inflation (P) as inflation is not a purely monetary phenomenon as shown in 1.

```python
# Creating a lag variable for CPI for all countries and adding it to the dataset.
merged_df['lag_CPI_pak'] = merged_df['CPI_pak'].shift()
merged_df['lag_CPI_ind'] = merged_df['CPI_ind'].shift()
merged_df['lag_CPI_sri'] = merged_df['CPI_sri'].shift()
merged_df['lag_CPI_bang'] = merged_df['CPI_bang'].shift()

merged_df.head()

```

```python
#Running the OLS for each country
##For Pakistan
X=merged_df[['lag_CPI_pak', 'M2_gr_pak', 'RGDP_gr_pak', 'IR_pak', 'SPW_pak', 'ER_pak']]
y=merged_df['CPI_pak']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X, missing='drop').fit()

# Print the summary of the model
print(model.summary())
```

The results indicate that the lag of CPI, money supply growth rate, interest rate, exchange rate, and support price of wheat all have a positive impact on inflation. Only GDP growth rate has a negative relationship in inflation which is in line with the theory.
(Note: only results for lag of CPI and SPW are signficant)

```python
# For India
X=merged_df[['lag_CPI_ind', 'M2_gr_ind', 'RGDP_gr_ind', 'SPW_ind', 'ER_ind']]
y=merged_df['CPI_ind']

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X, missing='drop').fit()

# Print the summary of the model
print(model.summary())

```
